```python
# 라이브러리 사용하기
import matplotlib.pyplot as plt
```

# matplotlib
- 2D 그래프로 시각화가 가능한 라이브러리
- 파이썬에서 데이터를 차트나 플롯으로 그려줌
- 차트의 종류는 산점도, 바, 파이, 히스토그램, 꺾은선 그래프 등 다양함

### 데이터 시각화란?
- 광범위하게 분산된 너무 많은 방대한 양의 자료를 한 눈에 볼 수 있도록 도표나 차트로 정리하는 것
- 시각화를 통해 데이터의 특징을 쉽게 파악할 수 있다.

### 기본적인 그래프 그려보기

```python
# X축과 Y축을 지정해서 선 그래프를 그려보자

x = [1,2,3,4,5]
y = [1,5,7,8,6]

# plot은 선 그래프를 그려주는 명령어
# plot(x축, y축, linestyle)
plt.plot(x,y,ls ='--')

#그래프를 출력
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

# lw(linewidth) --> 선 굵기
plt.plot(x,y,ls ='--', lw = 5)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

# marker = 표시 방법
plt.plot(x, y, ls = '--', lw = 5, marker = 'p')

plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,5,7,8,6]

# c : 선 색깔
plt.plot(x, y, ls = '--', lw = 5, marker = 'p', c = 'm')

plt.show()
```

```python
# numpy 라이브러리 로딩
import numpy as np
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

# mec : 마커 포인트 강조할 색상
plt.plot(x, y, ls = '--', lw = 5, marker = 'p', c = 'm', mec = 'b')

plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

# ms : 마커 크기
# mfc : 마커 채울 색상
plt.plot(x, y, ls = ':', marker = 'D', c = 'r', ms = 3, mfc = 'b')

plt.show()
```

### 연습문제

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls = '--', c ='b', lw =5, marker='o', ms=15, mec='g', mfc='r', mew=5)

plt.show()
```

### 차트 옵션

##### 차트 내부 그림의 범위를 지정
- lim()

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls = '--', c ='b', lw =5, marker='o', ms=15, mec='g', mfc='r', mew=5)

plt.xlim(-1, 10)
plt.ylim(-5, 12)

plt.show()
```

##### tic 설정
- 그래프에서 x축, y축 상의 위치 표시 지점을 틱 이라고 함.
- 틱은 일반적으로 matplotlib에서 자동으로 정해주지만(디폴트), 수동으로 설정도 가능

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls = '--', c ='b', lw =5, marker='o', ms=15, mec='g', mfc='r', mew=5)

plt.xticks([0,3,6])
plt.yticks([1,5,9])

plt.show()
```

##### 그리드
- 그래프에서 틱의 위치를 잘 보여주기 위해 그림 중간에 선이 자동으로 그려지는 것

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x, y, ls = '--', c ='b', lw =5, marker='o', ms=15, mec='g', mfc='r', mew=5)

plt.grid()

plt.show()
```

##### 여러개의 그래프를 한번에 그리기

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

x1 = np.arange(10)
y1 = np.arange(10)

# 첫 번째 플롯
plt.plot(x, y, ls = '--', c ='b', lw =5, marker='o', ms=15, mec='g', mfc='r', mew=5)

# 두 번째 플롯
plt.plot(x1,y1, ls=':')

plt.grid()
plt.show()
```

### Pie Chart(파이 차트)

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15,30,45,10]
colors = ['yellow', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.title('Pie Chart')

plt.pie(sizes, explode = explode, labels=labels, colors=colors, 
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.axis('equal')
plt.show()
```

```python
plt.pie(sizes, explode = explode, labels=labels, colors=colors)

plt.show()
```
