**1년 랭킹 가져오기**

```python
# pandas 활용하여 날짜 형성
date = pd.date_range(start='2022-01-01', end='2022-09-25')
date
```

```python
# 문자열 포매팅
days = date.strftime('%y%m%d')
days
```

```python
# 반복문 진행사항 확인할 수 있는 라이브러리
from tqdm import tqdm_notebook
```

```python
title_list = []
point_list = []
rank_list = []
date_list = []

for day in tqdm_notebook(days):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20' + day
    res = req.get(url)
    soup = bs(res.text,'lxml')

    title = soup.select('.tit5>a')
    point = soup.select('.point')

    for i in range(len(title)):
        title_list.append(title[i].text)
        point_list.append(point[i].text)
        rank_list.append(i+1)
        date_list.append(day)
```

```python
dic = {'순위':rank_list,'영화제목':title_list,'별점':point_list,'날짜':date_list}
df = pd.DataFrame(dic)
df = df.set_index('순위')
df
```

**네이버 평점 리뷰 가져오기**

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://movie.naver.com/movie/point/af/list.naver'
res = req.get(url)
res
```

```python
soup = bs(res.text, 'lxml')
soup
```

```python
# 영화제목
title =  soup.select('.movie')
title[2].text
```

```python
# 영화 평점
point = soup.select('.list_netizen_score>em')
point[2].text
```

```python
# 리뷰내용
# 불필요한 태그 삭제 .extract()
# 개행 삭제 .strip()

score = soup.select('.list_netizen_score') # 별점
report = soup.select('.report') # 신고

for i in title:
    i.extract()
for i in score:
    i.extract()
for i in report:
    i.extract()
    
content = soup.select('.title')
content[2].text.strip()
```

```python
title_list = []
point_list = []
content_list = []

for i in range(len(title)):
    title_list.append(title[i].text)
    point_list.append(point[i].text)
    content_list.append(content[i].text.strip())
```

```python
dic = {'영화제목':title_list,'리뷰':content_list,'별점':point_list}
df = pd.DataFrame(dic)
df = df.set_index('영화제목')
df
```

#### 10페이지까지 가져오기

```python
t_list = []
title_list = []
rate_list = []

for i in range(1, 11):
    url = 'https://movie.naver.com/movie/point/af/list.naver?&page=' + str(i)
    res = req.get(url)
    soup = bs(res.text, 'lxml')
    
    a = soup.select('a.movie.color_b')
    span = soup.select('td.title > div > span')
    em = soup.select('td.title > div > em')
    a2 = soup.select('a.report')
    
    for i in range(len(a)):
        a[i].extract()
        span[i].extract()
        em[i].extract()
        a2[i].extract()

    title = soup.select('td.title')
    
    for i in range(len(title)):
        t_list.append(a[i].text)
        title_list.append(title[i].text.strip())
        rate_list.append(em[i].text)
```

```python
dic = {'영화제목':t_list,'리뷰':title_list,'별점':rate_list}
```

```python
df = pd.DataFrame(dic)
df
```
