```python
from bs4 import BeautifulSoup as bs

from selenium import webdriver as wb # 웹을 통제하기 위한 라이브러리
from selenium.webdriver.common.keys import Keys # 웹에 키 값을 전달하기 위한 라이브러리
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
url = 'https://www.youtube.com/c/%EC%A7%80%EB%AC%B4%EB%B9%84/videos'

driver = wb.Chrome()
driver.get(url)
```

```python
body = driver.find_element(By.CSS_SELECTOR, 'body')
```

```python
for i in range(50):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.3)
```

```python
soup = bs(driver.page_source, 'lxml')
```

```python
title = soup.select('a#video-title')
```

```python
for i in title:
    print(i.text)
```

```python
count = soup.select('span.style-scope.ytd-grid-video-renderer:nth-child(1)')
```

```python
title_list = []
count_list = []

for i in range(len(title)):
    title_list.append(title[i].text)
    count_list.append(count[i].text)

dic = {'영상제목' : title_list, '조회수' : count_list}
df = pd.DataFrame(dic)
df
```

```python
re = []

for i in df['조회수'].values:
    re.append(i.replace("조회수 ", ""))
```

```python
df['조회수'] = re
```
