```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
url = 'https://corners.gmarket.co.kr/Bestsellers'

driver = wb.Chrome()
driver.get(url)
```

```python
# 버튼 누르기 위해 제어하려고 find_elements 사용
img = driver.find_elements(By.CSS_SELECTOR, 'img.lazy')
```

```python
img[0].click()
```

```python
# 컨텐츠를 가지고 오기 위해 파싱
soup = bs(driver.page_source, 'lxml')
soup
```

```python
# 상품
```

```python
# select 는 리스트 형(한개 가져올 때[0].text 이렇게 가져와야 해)인데 
# 여기서는 딱 한개만 가져올거니깐 select_one을 쓰면 됨
title = soup.select_one('h1.itemtit')
title
```

```python
# 가격
```

```python
price = soup.select_one('strong.price_real')
price
```

```python
# 원산지
```

```python
span = soup.select('span.sp_vipgroup')

for i in span:
    i.extract()
```

```python
orgin = soup.select('li.list-item-origin.uxeslide_item>div.box__information-title>div.box__information')
orgin[0].text.strip()
```

```python
# 페이지 뒤로가기
driver.back()
```

```python
# 이제 200개 전체적으로 가져오기(수업시간엔 10번까지만 함. 200번 너무 오래걸림)
```

```python
title_list = []
price_list = []
orgin_list = []

img = driver.find_elements(By.CSS_SELECTOR, 'img.lazy')

for i in range(10):
    img = driver.find_elements(By.CSS_SELECTOR, 'img.lazy')
    img[i].click()
    time.sleep(0.5)
    
    soup = bs(driver.page_source, 'lxml')
    title = soup.select_one('h1.itemtit')
    price = soup.select_one('strong.price_real')

    span = soup.select('span.sp_vipgroup')
    for i in span:
        i.extract()

    orgin = soup.select('li.list-item-origin.uxeslide_item > div.box__information-title > div.box__information')
    orgin[0].text.strip()
    
    title_list.append(title.text)
    price_list.append(price.text)
    orgin_list.append(orgin[0].text.strip())
    
    driver.back()
    time.sleep(0.5)
```

```python
print(len(title_list))
print(len(price_list))
print(len(orgin_list))
```

```python
import pandas as pd
```

```python
dic = {'메뉴명' : title_list, '가격' : price_list, '원산지':orgin_list}

df = pd.DataFrame(dic)
df
```
